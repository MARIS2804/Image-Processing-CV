```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("img1.jpg",0) 
alpha, beta = 1.5, 40
adj_img = cv2.convertScaleAbs(img, alpha=alpha, beta=beta)
plt.imshow(cv2.cvtColor(adj_img, cv2.COLOR_BGR2RGB))
plt.title("Brightness & Contrast Adjusted")
plt.show()
```

```python
kernel = np.array([[0, -1, 0],
                   [-1, 5, -1],
                   [0, -1, 0]])
sharp_img = cv2.filter2D(img, -1, kernel)
plt.imshow(cv2.cvtColor(sharp_img, cv2.COLOR_BGR2RGB))
plt.title("Sharpened Image")
plt.show()
```

```python
img = cv2.imread("img1.jpg",1)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
thresh, maxval = 127, 255
ret, binary = cv2.threshold(gray, thresh, maxval, cv2.THRESH_BINARY)
ret, binary_inv = cv2.threshold(gray, thresh, maxval, cv2.THRESH_BINARY_INV)
ret, trunc = cv2.threshold(gray, thresh, maxval, cv2.THRESH_TRUNC)
ret, tozero = cv2.threshold(gray, thresh, maxval, cv2.THRESH_TOZERO)
ret, tozero_inv = cv2.threshold(gray, thresh, maxval, cv2.THRESH_TOZERO_INV)
```

```python
for img_out, title in [(binary, "Binary Threshold"), (binary_inv, "Binary Inv Threshold"),
                      (trunc, "Trunc Threshold"), (tozero, "Tozero Threshold"), (tozero_inv, "Tozero Inv Threshold")]:
    plt.imshow(img_out, cmap="gray")
    plt.title(title)
    plt.axis("off")
    plt.show()
```

```python
maxval, block_size, C = 255, 11, 2
th_mean = cv2.adaptiveThreshold(gray, maxval, cv2.ADAPTIVE_THRESH_MEAN_C,
                                cv2.THRESH_BINARY, block_size, C)
th_gauss = cv2.adaptiveThreshold(gray, maxval, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                 cv2.THRESH_BINARY, block_size, C)
```

```python
plt.imshow(th_mean, cmap="gray")
plt.title("Adaptive Mean Threshold")
plt.axis("off")
plt.show()
plt.imshow(th_gauss, cmap="gray")
plt.title("Adaptive Gaussian Threshold")
plt.axis("off")
plt.show()
```

```python
ret, otsu = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
plt.imshow(otsu, cmap="gray")
plt.title("Otsu Thresholding")
plt.axis("off")
plt.show()
```
