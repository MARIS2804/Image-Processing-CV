```python
import matplotlib.pyplot as plt
```

```python
import cv2
import numpy as np

img1 = cv2.imread("img1.jpg")
img2 = cv2.imread("img2.jpg")

img2 = cv2.resize(img2, (img1.shape[1], img1.shape[0]))

img1_gray = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
img2_gray = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

and_img = cv2.bitwise_and(img1_gray, img2_gray)
or_img = cv2.bitwise_or(img1_gray, img2_gray)
xor_img = cv2.bitwise_xor(img1_gray, img2_gray)
```

```python
plt.imshow(and_img, cmap='gray')
plt.title('Bitwise AND')
plt.axis('off')
plt.show()

plt.imshow(or_img, cmap='gray')
plt.title('Bitwise OR')
plt.axis('off')
plt.show()

plt.imshow(xor_img, cmap='gray')
plt.title('Bitwise XOR')
plt.axis('off')
plt.show()
```

```python
avg_blur = cv2.blur(img1, (5, 5))
gauss_blur = cv2.GaussianBlur(img1, (5, 5), 0)
median_blur = cv2.medianBlur(img1, 5)

```

```python
plt.imshow(avg_blur, cmap='gray')
plt.title('Average Blur')
plt.show()

plt.imshow(gauss_blur, cmap='gray')
plt.title('Gaussian Blur')
plt.show()

plt.imshow(median_blur, cmap='gray')
plt.title('Median Blur')
plt.show()
```

```python
shrink = cv2.resize(img1, None, fx=0.5, fy=0.5, interpolation=cv2.INTER_AREA)
zoom_linear = cv2.resize(img1, None, fx=2, fy=2, interpolation=cv2.INTER_LINEAR)
zoom_cubic = cv2.resize(img1, None, fx=2, fy=2, interpolation=cv2.INTER_CUBIC)
```

```python
plt.imshow(shrink, cmap='gray')
plt.title('Shrink (INTER_AREA)')
plt.show()

plt.imshow(zoom_linear, cmap='gray')
plt.title('Zoom (INTER_LINEAR)')
plt.show()

plt.imshow(zoom_cubic, cmap='gray')
plt.title('Zoom (INTER_CUBIC)')
plt.show()
```
