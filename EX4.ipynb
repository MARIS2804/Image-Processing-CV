```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("img1.jpg")
template = cv2.imread("img2.jpg")

img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
template_gray = cv2.cvtColor(template, cv2.COLOR_BGR2GRAY)
eq_img = cv2.equalizeHist(img_gray)
```

```python
plt.imshow(img_gray, cmap="gray")
plt.title("Original Grayscale")
plt.axis("off")
plt.show()
plt.figure()
plt.hist(img_gray.ravel(), 256, [0,256])
plt.title("Original Histogram")
plt.xlabel("Pixel Value")
plt.ylabel("Frequency")
plt.show()

# Equalized Grayscale & Histogram
plt.imshow(eq_img, cmap="gray")
plt.title("Equalized Grayscale")
plt.axis("off")
plt.show()
plt.figure()
plt.hist(eq_img.ravel(), 256, [0,256])
plt.title("Equalized Histogram")
plt.xlabel("Pixel Value")
plt.ylabel("Frequency")
plt.show()
```

```python
src_hist, _ = np.histogram(img_gray.flatten(), 256, [0,256])
tmpl_hist, _ = np.histogram(template_gray.flatten(), 256, [0,256])
src_cdf = np.cumsum(src_hist) / src_hist.sum()
tmpl_cdf = np.cumsum(tmpl_hist) / tmpl_hist.sum()
mapping = np.array([np.argmin(np.abs(tmpl_cdf - v)) for v in src_cdf])
matched = mapping[img_gray]
```

```python
plt.imshow(template_gray, cmap="gray")
plt.title("Template Grayscale")
plt.axis("off")
plt.show()
plt.figure()
plt.hist(template_gray.ravel(), 256, [0,256])
plt.title("Template Histogram")
plt.xlabel("Pixel Value")
plt.ylabel("Frequency")
plt.show()

# Histogram Matched & Histogram
plt.imshow(matched, cmap="gray")
plt.title("Histogram Matched")
plt.axis("off")
plt.show()
plt.figure()
plt.hist(matched.ravel(), 256, [0,256])
plt.title("Matched Histogram")
plt.xlabel("Pixel Value")
plt.ylabel("Frequency")
plt.show()
```
